#### Performing clustering 

```python
data = df.drop(['name','id','total scores'], axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
gender_enc = LabelEncoder()
```

```python
career_aspiration = LabelEncoder()
```

```python
data['gender'] = gender_enc.fit_transform(data['gender'])
```

```python
data['career_aspiration'] = career_aspiration.fit_transform(data['career_aspiration'])
```

```python
data.head(2)
```

```python
'performing elbow method'
from sklearn.cluster import KMeans
```

```python
K = range(1,11)
inertia = []
mapping_inertia = {}
```

```python
for k in K:
    kmeans_model = KMeans( n_clusters = k , random_state = 23 )
    kmeans_model.fit(data)
    inertia.append(kmeans_model.inertia_)
    mapping_inertia[k] = inertia[-1]
```

```python
sns.lineplot(x = K, y = inertia, marker = '*', color = 'blue')
plt.show()
```

```python
k_model = KMeans(n_clusters = 3)
data['clusters'] = k_model.fit_predict(data)
```

```python
data.groupby('clusters')[subjects].mean().mean(axis = 1)
```

```python
mapping_clusters = {2:'high_performance', 1:'average_student', 0:'low_performance'}
```

```python
df['performance categories'] = data['clusters'].map(mapping_clusters)
```

#### Hypothesis Testing 

```python
'''taking 30 samples and checking their mean'''
male_mean = df[df.gender == 'male']['math_score'].sample(n = 30, random_state = 21).mean()
female_mean = df[df.gender == 'female']['math_score'].sample(n = 30, random_state  =21).mean()
male_std = df[df.gender == 'male']['math_score'].sample(n = 30, random_state = 21).std()
female_std = df[df.gender == 'female']['math_score'].sample(n = 30, random_state  =21).std()
n_m = 30
n_f = 30
```

```python
import scipy.stats as stats
alpha = 0.05
degree_of_freedom = (n_m + n_f) - 2
t_stats = np.abs((male_mean - female_mean)/np.sqrt((male_std**2/n_m) + (female_std**2/n_f))) # computing the t statisitics 
p_value = stats.t.cdf(-t_stats, degree_of_freedom) # computing the p-value
critical_value = stats.t.ppf((1 - alpha)/2, degree_of_freedom) # computing critical value
print('p value:', p_value)
print('t statisitics:', t_stats)
print('critical value:', critical_value)
if t_stats > critical_value :
    print('The test is significant and we reject the Null Hypothesis')
else:
    print('The test is not significant and we fail to reject the null hypothesis')
```

```python
df.career_aspiration.value_counts().head(5)
'''we are selecting the first four common groups'''
science_subjects = ['math_score', 'physics_score', 'chemistry_score', 'biology_score']
software_engineer = df[df.career_aspiration == 'Software Engineer'][science_subjects].sample(n = 40, random_state = 12)
business_owner = df[df.career_aspiration == 'Business Owner'][science_subjects].sample(n = 40, random_state = 12)
banker = df[df.career_aspiration == 'Banker'][science_subjects].sample(n = 40, random_state = 12)
lawyer = df[df.career_aspiration == 'Lawyer'][science_subjects].sample(n = 40, random_state = 12)
```

```python
from scipy import stats
f_statistics, pvalue = stats.f_oneway(software_engineer, business_owner, banker, lawyer)
print
for p, subjects in zip(pvalue, science_subjects):
    if p < 0.05 :
        print(f'The test is significant and career aspiration is significantly associated with {subjects} ')
    else:
        print(f'The test is insignificant and career aspiration is not significantly associated with {subjects}')
```
